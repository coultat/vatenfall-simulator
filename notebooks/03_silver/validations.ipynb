### Check the existence of the silver_tables

```python
from src.utils.load_config import paths_config as pc
import logging


tables = [t for t in pc['tables']['silver'].values()]
tables.extend([reference_t for reference_t in pc['tables']['references'].values()])
```

```python
tables
```

```python
for table in tables:
    try:
        query = f"DESCRIBE TABLE {table}"
        spark.sql(query)
    except Exception as e:
        logging.error(f"The table {table} was not found")
```

```python
for table in tables:
    df = spark.read.table(table)
    assert df.count() > 0, f"The table {table} is empty"
```
