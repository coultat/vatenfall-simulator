```python
from src.utils.load_config import paths_config as pc
from pyspark.sql.functions import max as spark_max
from pyspark.sql.functions import round as spark_round
from pyspark.sql.functions import sum as spark_sum
from pyspark.sql.functions import avg, col, when
from pyspark.sql import DataFrame


def get_gold_daily_market_summary(df: DataFrame) -> DataFrame:
 return  ( df
          .groupBy("event_date", "region")
          .agg(spark_max(col('price_eur_mwh')).alias('max_day_price_mwh'),
               spark_round(avg(col('price_eur_mwh')), 2).alias('avg_price_eur_mwh'),
               spark_sum(col('volume_mwh')).alias('total_volume'),
               spark_sum(when(col("price_category") == 'High', 1)
                         .otherwise(0)).alias('price_volume')
               )
          .orderBy(col('event_date').asc())
          
)
```
