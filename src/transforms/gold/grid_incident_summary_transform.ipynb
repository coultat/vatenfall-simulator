```python
from pyspark.sql.functions import round as spark_round
from pyspark.sql.functions import sum as spark_sum
from pyspark.sql.functions import count as spark_count
from pyspark.sql.functions import col, avg
from pyspark.sql import DataFrame


def get_gold_grid_incident_summary(df: DataFrame) -> DataFrame:
    return ( df
            .groupBy(col('event_date'), col('region'), col('severity'))
            .agg(
                spark_count(col('severity')).alias('severity_cases'),
                spark_round(spark_sum(col('duration_minutes')), 2).alias('total_of_status'),
                spark_round(avg(col('duration_minutes')), 2).alias('average_duration')
            )
            .orderBy(col('event_date').asc())
    )
```
