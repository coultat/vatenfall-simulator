```python
from pyspark.sql import DataFrame
from pyspark.sql.functions import col

def get_gold_regional_operations_dashboard(daily_market_df: DataFrame, grid_incident_df: DataFrame, weather_impact_df: DataFrame) -> DataFrame:
    return ( daily_market_df.alias('daily_market')
            .join(
                grid_incident_df.alias('grid_incident'),
                (col('daily_market.event_date') == col('grid_incident.event_date')) &
                (col('daily_market.region') == col('grid_incident.region'))
            )
            .join(weather_impact_df.alias('weather_impact'),
                (col('weather_impact.weather_event_date') == col('daily_market.event_date')) &
                (col('weather_impact.weather_region') == col('daily_market.region'))
                )
            .drop(col('grid_incident.region'), col('weather_impact.weather_region'), col('grid_incident.event_date'))
            .orderBy(col('daily_market.event_date').asc())
            )
```
