```python
from pyspark.sql import DataFrame
from pyspark.sql.functions import col


def get_silver_regional_operations_base(grid_events_df: DataFrame, asset_reference_df: DataFrame, weather_events_df: DataFrame) -> DataFrame:
      return  (grid_events_df
          .alias('grid_events')
          .join(
              asset_reference_df.alias('asset_reference')
              .withColumnRenamed('region', 'asset_region'),
              col('grid_events.asset_id') == col('asset_reference.asset_id')
          )
          .join(
              weather_events_df.alias('weather'),
              (col('grid_events.region') == col('weather_region')) & 
              (col('grid_events.event_date') == col('weather.weather_event_date'))
          )
          .drop(col('asset_reference.asset_id'), col('asset_reference.asset_name'))
      )
```
