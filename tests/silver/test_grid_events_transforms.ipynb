```python
import pytest
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, TimestampType
from datetime import datetime


def test_normalize_grid_events():

    test_data = [
        ("2024-01-01 10:30:00", "  Madrid  ", "  ERROR  ", "  HIGH  ", "60", "rescued", datetime(2024,1,1,10,30,0)),
    ]
    
    schema = StructType([
        StructField("last_updated_ts", StringType()),
        StructField("region", StringType()),
        StructField("event_type", StringType()),
        StructField("severity", StringType()),
        StructField("duration_minutes", StringType()),
        StructField("_rescued_data", StringType()),
        StructField("event_ts", TimestampType()),
    ])
    
    df = spark.createDataFrame(test_data, schema)  
    result = normalize_grid_events(df)
    

    assert "event_ts" not in result.columns, "event_ts should have been dropped"
    assert "_rescued_data" not in result.columns, "_rescued_data should have been dropped"
    assert result.schema["duration_minutes"].dataType.typeName() == "integer"
    assert result.schema["last_updated_ts"].dataType.typeName() == "timestamp"
    assert result.schema["event_date"].dataType.typeName() == "date"
    

    assert result.collect()[0]["region"] == "Madrid"
    
    print("✅ test_normalize_grid_events pasado")

test_normalize_grid_events()
```
